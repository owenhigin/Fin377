With hackers becoming more and more savvy, it is becoming increasingly important to use secure and unique passwords. Unfortunately, many people still do not use secure passwords. Did you know that the passwords “123456” and “password” were still the most used passwords of 2021? And that it takes a hacker less than 1 second to crack them?

To enforce their users to use secure passwords, many websites make use of a password validator nowadays. This is a small tool that checks if a password is secure enough and does not contain any common keywords. In this assignment, we will create a very basic password validator by combining all the things we learned in the past assignments!

Let's write code to prompt the user to fill in their password and will do the following basic checks on it:

- The password may not contain the keywords “password” or “123456”.
- The password may not start with the keyword “qwerty” but it may contain it.
- The password must contain at least one non-alphabetic character.
- The password must contain at least one non-digit character.

After analyzing the password, you will print the conclusion. If the password did not pass the tests, executed in the order above, it will print the error message as you can find in the list above.
````python
password = input("Your password: ")

# work here. then test it with different passwords!

if "password" in password or "123456" in password:
    print('The password may not contain the keywords "password" or "123456".')
elif password.startswith("qwerty"):
    print('The password may not start with the keyword "qwerty" but it may contain it.')
elif password.isalpha():
    print('The password must contain at least one non-alphabetic character.')
elif password.isdigit():
    print('The password must contain at least one non-digit character.')
elif '@' not in password and '$' not in password and '!' not in password:
    print("The password must contain at least one of the following special characters: $, @ or !.")
elif password.isupper() or password.islower():
    print("The password must have both uppercase and lowercase letters.")
else:
    print("The password is valid!")
````

## Q2: Turn that into a function

If your code above is just a conditional check, turn it into a function. It should accept a string argument and output the error or success message above. These should be `return`ed, not printed.

````python
def valid_password(password: str):
    #work here
    if "password" in password or "123456" in password:
        return 'The password may not contain the keywords "password" or "123456".'
    elif password.startswith("qwerty"):
        return 'The password may not start with the keyword "qwerty" but it may contain it.'
    elif password.isalpha():
        return 'The password must contain at least one non-alphabetic character.'
    elif password.isdigit():
        return 'The password must contain at least one non-digit character.'
    elif '@' not in password and '$' not in password and '!' not in password:
        return "The password must contain at least one of the following special characters: $, @ or !."
    elif password.isupper() or password.islower():
        return "The password must have both uppercase and lowercase letters."
    else:
        return "The password is valid!"

# leave these. If your code works, this cell will run without producing an error
assert valid_password('123456') == 'The password may not contain the keywords "password" or "123456".'
assert valid_password('I am the password') == 'The password may not contain the keywords "password" or "123456".'
assert valid_password('qwerty') == 'The password may not start with the keyword "qwerty" but it may contain it.'
assert valid_password('Hiqwerty') == 'The password must contain at least one non-alphabetic character.'
assert valid_password('1233874356') == 'The password must contain at least one non-digit character.'
assert valid_password('HI qwerty 1') == 'The password must contain at least one of the following special characters: $, @ or !.'
assert valid_password('hi qwerty 1!') == 'The password must have both uppercase and lowercase letters.'
assert valid_password('HI qwerty 1!') == 'The password is valid!'
````
